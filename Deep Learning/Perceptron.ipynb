# load ex1 using tf.keras


```python
# adaline using keras
from tensorflow.keras import models, layers
import numpy as np

# AND 

# INPUT / OUTPUT DATA
# in
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# out
T = np.array([0,0,0,1])

#   .
model = models.Sequential()

#   . (unit :  ( ), input_shape :  , activation : )
model.add(layers.Dense(units = 1,  activation='linear',input_shape=(2,)))

#  sigmoid  ,   : (0 ,1   ),  .
model.add(layers.Dense(units = 1,  activation='sigmoid',input_shape=(2,)))



# training( ) optimizer loss     .
model.compile(optimizer='sgd', # optimizer cost   (sgd : ) 
                              #    ,  . adam   
             loss="mse", # loss cost  .(cost func y = Wx + b  )(loss = target, y )
            loss="binary_crossentropy", #     ,   sigmoid - binary_crossentropy !!!
             metrics = ['acc'])


# epochs( ) = 
# loss    
model.fit(X,T,epochs = 10000, batch_size = 1, verbose = 1)
```

```python
# test,  test   .
Xtest_loss, Xtest_acc = model.evaluate(X,T)
print(Xtest_acc)
```

```python
# accuracy 
Ttest = model.predict(X)
                
print(Ttest) # 0 0 0 1
            # 0 0 / 0 1 / 1 0 / 1 1  
```

# IRIS Data set

##### 2) Train/Test

```python
# adaline using keras
from tensorflow.keras import models, layers
import numpy as np
from sklearn.datasets import load_iris

iris_data = load_iris()

X = iris_data.data
y = iris_data.target
```

```python
# 
model = models.Sequential()
model.add(layers.Dense(units = 1,  activation='linear',input_shape=(4,)))

# 
model.compile(optimizer='sgd',
             loss="mse",
             metrics = ['acc'])

model.fit(X,y,epochs = 1000, batch_size = 1, verbose = 1)
```

```python
# test
Xtest_loss,Xtest_acc = model.evaluate(X,y)
print(Xtest_acc)
```

```python
# accuracy 
Ttest = model.predict(X)
print(Ttest) 
```
